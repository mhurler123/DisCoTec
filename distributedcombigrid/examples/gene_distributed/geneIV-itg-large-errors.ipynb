```python
def calcEVerror(refdir1, refdir2, combidir):
    print combidir, refdir1, refdir2
    
    # read ref1 lambda from omega.dat
    for line in open(refdir1 + '/omega.dat'):
        tmp = line.split()
        ref_lambda1 = complex(float(tmp[1]),float(tmp[2]))
        
    for line in open(refdir2 + '/omega.dat'):
        tmp = line.split()
        ref_lambda2 = complex(float(tmp[1]),float(tmp[2]))
        
    levels, lambdas, combi_lambda = getEVcombi(combidir)
    
    err1_lambdas = []
    err2_lambdas = []
    
    for i in range(0,len(levels)):
        err1_lambdas.append( abs(ref_lambda1 - lambdas[i]) )
        err2_lambdas.append( abs(ref_lambda2 - lambdas[i]) )
        print str(levels[i][:-1]) + ' ' + str( lambdas[i]) + ' ' + str( abs(ref_lambda1 - lambdas[i]) ) + ' ' + str( abs(ref_lambda2 - lambdas[i]) )
        
    min_err1 = min( err1_lambdas )
    min_err2 = min( err2_lambdas )
        
    print 'combi ' + str(combi_lambda) + ' ' + str( abs(ref_lambda1 - combi_lambda ) )  + ' ' + str( abs(ref_lambda2 - combi_lambda ) ) 
    print 'ref1 ' + str(ref_lambda1)
    print 'ref2 ' + str(ref_lambda2)
    
    err1_combi = abs(ref_lambda1 - combi_lambda )
    err2_combi = abs(ref_lambda2 - combi_lambda )
    
    return err1_combi, err2_combi, min_err1, min_err2

def getRefEV(dir):
    for line in open( dir + '/omega.dat' ):
        return complex( float(line.split()[1]), float(line.split()[2]) )
```

```python
def getRuntime( dir ):
    filename = dir + '/ginstance/times.dat'
    for line in open(filename):
        if line.split()[0] == 'runfirst':
            return float( line.split()[1] )
        
    return -1.0

def getProcs( dir ):
    filename = dir + '/ctparam'
    for line in open(filename):
        if len( line.split() ) < 1:
            continue
        
        if line.split()[0] == 'nprocs':
            return float( line.split()[2] )
        
    return -1.0

def getResources( dir ):
    return getRuntime(dir) * getProcs(dir) / 3600.
```

```python
def getResourceVsErrorCombi( errbase, resbase, refdir, combidir ):
    _, _, evcombi = getEVcombi( combidir )
    errcombi = abs( getRefEV(refdir) - evcombi )
    rescombi = getResources(combidir)
    
    return abs( errbase - errcombi ) / abs( resbase - rescombi )
```

```python
def readIntervall(file):
    x = []
    y = []
    for line in open(file):
        x.append( float( line.split()[0] ) )
        y.append( float( line.split()[1] ) )
    
    return x,y
```
