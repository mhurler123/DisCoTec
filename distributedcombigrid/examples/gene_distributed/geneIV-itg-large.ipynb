```python
coordDict = {'w':8,'s':0,'y':0}
vlineplot = 16
x_lim = (0,256)
```

```python
def plotCheckpoint(folder, normalize=False):
    cp = checkpoint2Local(folder)
    cp.createAllData()
    pdf = GenePdf(cp.data)
    
    if normalize:
        # normalize abs
        data = cp.data / np.linalg.norm(cp.data.flatten(),2)
        
        pdf = GenePdf(data)

    figure(figsize=(20,4))
    subplot(141)
    d = (pdf.getBallooningSlice(coordDict)).real
    pcolormesh(d)
    xlim(x_lim)
    title('real')
    colorbar()
    
    subplot(142)
    d = (pdf.getBallooningSlice(coordDict)).imag
    pcolormesh(d)
    xlim(x_lim)
    title('imag')
    colorbar()
    
    subplot(143)
    #d = log10(abs(pdf.getBallooningSlice(coordDict)))
    d = abs(pdf.getBallooningSlice(coordDict))
    pcolormesh(d)
    xlim(x_lim)
    title('abs')
    colorbar()
    
    subplot(144)
    d = np.angle(pdf.getBallooningSlice(coordDict))
    pcolormesh(d)
    xlim(x_lim)
    title('angle')
    colorbar()
    
def calcErrorAbs(diagDir1, diagDir2):
    cp1 = checkpoint2Local(diagDir1)
    cp1.createAllData()
    pdf1 = GenePdf( cp1.data / np.linalg.norm(cp1.data.flatten(),2) )

    cp2 = checkpoint2Local(diagDir2)
    cp2.createAllData()
    pdf2 = GenePdf(cp2.data / np.linalg.norm(cp2.data.flatten(),2) )

    print 'L2 Error of Absolute = ' + str( ErrorNorms.L2Error(abs(pdf2.data),abs(pdf1.data)) )
    
## hier tests
class FullGrid:
    def __init__(self, filename):
        f = open(filename, "r")
        lines = f.readlines()

        # read dimension
        self.dim = int( lines[0] )

        # read extents
        self.ext = [int(x) for x in lines[1].split()]
        #print self.ext

        # read data
        size = 0
        datatmp = []
        for l in lines[2:]:
            tmp = l.replace("(",'').replace(")",'')
            real = float( tmp.split(",")[0] )
            imag = float( tmp.split(",")[1] )
            datatmp.append( complex(real,imag) )
            size += 1

        #print size
        
        # convert to multi dimensional array
        self.data = np.asarray( datatmp )
        self.data = self.data.reshape( (self.ext[::-1]) )
        print self.data.shape
        
        
def plotFullGrid(fg, normalize=False):
    data = fg.data[:,:-1,:-1,:-1,:,:-1]
    pdf = GenePdf(data)
    
    if normalize:
        # normalize abs
        data = data / np.linalg.norm(data.flatten(),2)
        
        pdf = GenePdf(data)

    figure(figsize=(20,4))
    subplot(141)
    d = (pdf.getBallooningSlice(coordDict)).real
    pcolormesh(d)
    xlim(x_lim)
    title('real')
    colorbar()
    subplot(142)
    d = (pdf.getBallooningSlice(coordDict)).imag
    pcolormesh(d)
    xlim(x_lim)
    title('imag')
    colorbar()
    subplot(143)
    #d = log10(abs(pdf.getBallooningSlice(coordDict)))
    d = abs(pdf.getBallooningSlice(coordDict))
    pcolormesh(d)
    xlim(x_lim)
    title('abs')
    colorbar()
    subplot(144)
    d = np.angle(pdf.getBallooningSlice(coordDict))
    pcolormesh(d)
    xlim(x_lim)
    title('angle')
    colorbar()
    
    
def calcErrorAbsFG_CP(fg, diagDir2):
    data1 = fg.data[:,:-1,:-1,:-1,:,:-1]
    pdf1 = GenePdf( data1 / np.linalg.norm(data1.flatten(),2) )

    cp2 = checkpoint2Local(diagDir2)
    cp2.createAllData()
    pdf2 = GenePdf(cp2.data / np.linalg.norm(cp2.data.flatten(),2) )

    print 'L2 Error of Absolute = ' + str( ErrorNorms.L2Error(abs(pdf2.data),abs(pdf1.data)) )
```
