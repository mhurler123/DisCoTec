```python
def plotCheckpoint(folder, normalize=False, mytitle=''):
    cp = checkpoint2Local(folder)
    cp.createAllData()
    pdf = GenePdf(cp.data)
    
    if normalize:
        # normalize abs
        data = cp.data / np.linalg.norm(cp.data.flatten(),2)
        
        pdf = GenePdf(data)
    
    '''
    subplot(141)
    d = (pdf.getBallooningSlice(coordDict)).real
    pcolormesh(d)
    xlim(x_lim)
    title('real')
    colorbar()
    
    subplot(142)
    d = (pdf.getBallooningSlice(coordDict)).imag
    pcolormesh(d)
    xlim(x_lim)
    title('imag')
    colorbar()
    '''
    
    #d = log10(abs(pdf.getBallooningSlice(coordDict)))
    d = abs(pdf.getBallooningSlice(coordDict))
    pcolormesh(d)
    xlim(x_lim)
    title(mytitle)
    colorbar()
    xlabel('xz')
    ylabel('v')
    '''
    subplot(144)
    d = np.angle(pdf.getBallooningSlice(coordDict))
    pcolormesh(d)
    xlim(x_lim)
    title('angle')
    colorbar()
    '''
    
def calcErrorAbs(diagDir1, diagDir2):
    cp1 = checkpoint2Local(diagDir1)
    cp1.createAllData()
    pdf1 = GenePdf( cp1.data / np.linalg.norm(cp1.data.flatten(),2) )

    cp2 = checkpoint2Local(diagDir2)
    cp2.createAllData()
    pdf2 = GenePdf(cp2.data / np.linalg.norm(cp2.data.flatten(),2) )

    print 'L2 Error of Absolute = ' + str( ErrorNorms.L2Error(abs(pdf2.data),abs(pdf1.data)) )
    
## hier tests
class FullGrid:
    def __init__(self, filename, parallel=False):
        if(parallel):
            f = open(filename, "r")
            
            # first 4 bytes -> dim
            self.dim = struct.unpack("i", f.read(4))[0]
            
            # next dim*4 bytes -> extents
            self.ext = []
            for i in range(self.dim):
                self.ext.append( struct.unpack("i", f.read(4))[0] )
                
            # read data
            length = np.multiply.accumulate(self.ext)[-1]
            print length
            stringData = f.read(16*length)
            self.data=np.fromstring(stringData,dtype=complex)
            
            self.data = self.data.reshape( (self.ext[::-1]) )
            
            print self.dim
            print self.ext
            print self.data.shape
            
        else:
            f = open(filename, "r")
            lines = f.readlines()

            # read dimension
            self.dim = int( lines[0] )

            # read extents
            self.ext = [int(x) for x in lines[1].split()]
            #print self.ext

            # read data
            size = 0
            datatmp = []
            for l in lines[2:]:
                tmp = l.replace("(",'').replace(")",'')
                real = float( tmp.split(",")[0] )
                imag = float( tmp.split(",")[1] )
                datatmp.append( complex(real,imag) )
                size += 1

            #print size

            # convert to multi dimensional array
            self.data = np.asarray( datatmp )
            self.data = self.data.reshape( (self.ext[::-1]) )
            print self.data.shape
        
        
def plotFullGrid(fg, normalize=False, mytitle=''):
    data = fg.data[:,:-1,:-1,:-1,:,:-1]
    pdf = GenePdf(data)
    
    if normalize:
        # normalize abs
        data = data / np.linalg.norm(data.flatten(),2)
        
        pdf = GenePdf(data)

    #figure(figsize=(4,4))
    '''
    subplot(141)
    d = (pdf.getBallooningSlice(coordDict)).real
    pcolormesh(d)
    xlim(x_lim)
    title('real')
    colorbar()
    subplot(142)
    d = (pdf.getBallooningSlice(coordDict)).imag
    pcolormesh(d)
    xlim(x_lim)
    title('imag')
    colorbar()
    '''
    #subplot(143)
    #d = log10(abs(pdf.getBallooningSlice(coordDict)))
    d = abs(pdf.getBallooningSlice(coordDict))
    pcolormesh(d)
    xlim(x_lim)
    title(mytitle)
    colorbar()
    xlabel('xz')
    ylabel('v')
    '''
    subplot(144)
    d = np.angle(pdf.getBallooningSlice(coordDict))
    pcolormesh(d)
    xlim(x_lim)
    title('angle')
    colorbar()
    '''
    
    
def calcErrorAbsFG_CP(fg, diagDir2):
    data1 = fg.data[:,:-1,:-1,:-1,:,:-1]
    pdf1 = GenePdf( data1 / np.linalg.norm(data1.flatten(),1) )

    cp2 = checkpoint2Local(diagDir2)
    cp2.createAllData()
    pdf2 = GenePdf(cp2.data / np.linalg.norm(cp2.data.flatten(),1) )

    print 'L2 Error of Absolute = ' + str( ErrorNorms.L1Error(abs(pdf2.data),abs(pdf1.data)) )
    
def calcErrorAbsFG_FG(fg1, fg2):
    data1 = fg1.data[:,:-1,:-1,:-1,:,:-1]
    data1 = data1 / np.linalg.norm(data1.flatten(),1)

    data2 = fg2.data[:,:-1,:-1,:-1,:,:-1]
    data2 = data2 / np.linalg.norm(data2.flatten(),1)

    print 'L2 Error of Absolute = ' + str( ErrorNorms.L1Error(abs(data1),abs(data2)) )
    
def calcErrorAbsSlicesFG_FG_w(fg1, fg2):
    data1 = fg1.data[:,:-1,:-1,:-1,:,:-1]
    data1 = data1 / np.linalg.norm(data1.flatten(),2)

    data2 = fg2.data[:,:-1,:-1,:-1,:,:-1]
    data2 = data2 / np.linalg.norm(data2.flatten(),2)
    
    nw = data1.shape[1]
    
    err = []
    nrm1 = []
    nrm2 = []
    
    for i in range(0,nw):
        tmp = ErrorNorms.L2Error(abs(data1[:,i,:,:,:,:]),abs(data2[:,i,:,:,:,:]))
        tmp2 = np.linalg.norm( abs(data1[:,i,:,:,:,:]).flatten(), 2)
        tmp3 = np.linalg.norm( abs(data2[:,i,:,:,:,:]).flatten(), 2)

        err.append(tmp)
        nrm1.append(tmp2)
        nrm2.append(tmp3)
    
    return err,nrm1,nrm2


def calcErrorAbsSlicesFG_FG_z(fg1, fg2):
    data1 = fg1.data[:,:-1,:-1,:-1,:,:-1]
    data1 = data1 / np.linalg.norm(data1.flatten(),2)

    data2 = fg2.data[:,:-1,:-1,:-1,:,:-1]
    data2 = data2 / np.linalg.norm(data2.flatten(),2)
    
    nw = data1.shape[3]
    
    err = []
    nrm1 = []
    nrm2 = []
    
    for i in range(0,nw):
        tmp = ErrorNorms.L2Error(abs(data1[:,:,:,i,:,:]),abs(data2[:,:,:,i,:,:]))
        tmp2 = np.linalg.norm( abs(data1[:,:,:,i,:,:]).flatten(), 2)
        tmp3 = np.linalg.norm( abs(data2[:,:,:,i,:,:]).flatten(), 2)

        err.append(tmp)
        nrm1.append(tmp2)
        nrm2.append(tmp3)
    
    return err,nrm1,nrm2


def calcErrorAbsSlicesFG_FG_x(fg1, fg2):
    data1 = fg1.data[:,:-1,:-1,:-1,:,:-1]
    data1 = data1 / np.linalg.norm(data1.flatten(),2)

    data2 = fg2.data[:,:-1,:-1,:-1,:,:-1]
    data2 = data2 / np.linalg.norm(data2.flatten(),2)
    
    nw = data1.shape[5]
    
    err = []
    nrm1 = []
    nrm2 = []
    
    for i in range(0,nw):
        tmp = ErrorNorms.L2Error(abs(data1[:,:,:,:,:,i]),abs(data2[:,:,:,:,:,i]))
        tmp2 = np.linalg.norm( abs(data1[:,:,:,:,:,i]).flatten(), 2)
        tmp3 = np.linalg.norm( abs(data2[:,:,:,:,:,i]).flatten(), 2)

        err.append(tmp)
        nrm1.append(tmp2)
        nrm2.append(tmp3)
    
    return err,nrm1,nrm2
    
    
def getEVcombi(dir):
    # read level vectors and coefficients from spaces dat
    levels = []
    coeffs = []
    for line in open(dir + 'ginstance/spaces.dat'):
        tmp = line.split()
        levels.append( [ int(x) for x in tmp[1:7] ] )
        coeffs.append(float(tmp[7]))
    
    #print levels
    #print coeffs
    
    # read omega.dat files
    lambdas = []
    for i in range(0, len(coeffs) ):
        for line in open(dir + '/ginstance' + str(i) + '/omega.dat'):
            tmp = line.split()
            lambdas.append( complex(float(tmp[1]),float(tmp[2])) )
            
    assert len(lambdas) == len(coeffs)
    
    #print lambdas
    
    # compute combi lambda
    combi_lambda = complex(0,0)
    for i in range(0, len(lambdas)):
        combi_lambda += coeffs[i] * lambdas[i]
        
    #print combi_lambda
    
    return levels, lambdas, combi_lambda

def calcEVerror(refdir, combidir):
    # read ref lambda from omega.dat
    for line in open(refdir + '/omega.dat'):
        tmp = line.split()
        ref_lambda = complex(float(tmp[1]),float(tmp[2]))
        
    levels, lambdas, combi_lambda = getEVcombi(combidir)
    
    for i in range(0,len(levels)):
        print str(levels[i][:-1]) + ' ' + str( lambdas[i]) + ' ' + str( abs(ref_lambda - lambdas[i]) )
        
    print 'combi ' + str(combi_lambda) + ' ' + str( abs(ref_lambda - combi_lambda ) )
    print 'ref ' + str(ref_lambda)
```
